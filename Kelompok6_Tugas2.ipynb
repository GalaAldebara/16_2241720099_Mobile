```python
import cv2 as cv
from google.colab.patches import cv2_imshow
import numpy as np

image = cv.imread('/content/drive/MyDrive/Images/KTP.jpeg')
def draw_zigzag(image, x1, y1, width, height, num_zigs, color, thickness):
    zigzag_height = height // 2

    for i in range(num_zigs):
        x_start = x1 + i * (width // num_zigs)
        x_end = x1 + (i + 1) * (width // num_zigs)
        if i % 2 == 0:
            cv.line(image, (x_start, y1), (x_end, y1 + zigzag_height), color, thickness = 15)
        else:
            cv.line(image, (x_start, y1 + zigzag_height), (x_end, y1), color, thickness = 15)

x1, y1 = 190, 321
width, height = 350, 25
num_zigs = 20
color = (255, 255, 0)
thickness = 5

draw_zigzag(image, x1, y1, width, height, num_zigs, color, thickness)

x2, y2 = 190, 348
width2, height2 = 90, 20
num_zigs2 = 10
color2 = (0, 128, 0)

draw_zigzag(image, x2, y2, width2, height2, num_zigs2, color2, thickness)
cv2_imshow(image)
```
